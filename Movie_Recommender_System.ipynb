```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert(movies.iloc[0]['genres'])
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0]['cast']
```

```python
def convert_cast(obj):
    L=[]
    ctr=0
    for i in ast.literal_eval(obj):
        if ctr!=3:
            L.append(i['name'])
            ctr=ctr+1
        else:
            break
    return L
```

```python
convert_cast(movies.iloc[0]['cast'])
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
movies.iloc[0]['crew']
```

```python
def get_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
get_director(movies.iloc[0]['crew'])
```

```python
movies['crew']=movies['crew'].apply(get_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x : x.split())
```

```python
movies.head()
```

We need to combine the names without any space between them so that the the words together become  one entity

```python
movies['genres']=movies['genres'].apply(lambda x :[i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x :[i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x :[i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x :[i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id','title','tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize('wolves')
```

```python
def preprocess(text):
    y=[]
    for i in text.split():
        y.append(lemmatizer.lemmatize(i))
    return ' '.join(y)
```

```python
new_df.to_csv('pre-processed movies data.csv',index=False)
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(preprocess)
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

Euclidean distance is not reliable for higher dimensions.Instead we use cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarities[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarities[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.loc[i[0]]['title'])
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
